### 2. Nutrition

This dataset contains multiple csv files with complex structure. Preprocessings are necessary.

```python
brnd_food = pd.read_csv(os.path.join(nutrition_path, "branded_food.csv"))
brnd_food = brnd_food.drop(columns = ['data_source', 'modified_date', 'available_date', 'discontinued_date', 'market_country'])
brnd_food.iloc[np.random.choice(brnd_food.index, 5, replace = False)]
```

```python
brnd_food.branded_food_category.nunique()
```

```python
brnd_food['fdc_id'].nunique()
```

```python
nutrient = pd.read_csv(os.path.join(nutrition_path, "nutrient.csv"))
nutrient.head()
```

```python
food_ntrn = pd.read_csv(os.path.join(nutrition_path, "food_nutrient.csv"))
food_ntrn.sort_values(by = 'fdc_id')
```

```python
food_ntrn['fdc_id'].nunique()
```

```python
brnd_ntrn = pd.merge(food_ntrn, brnd_food, on = 'fdc_id').drop(columns = ['id', 'data_points', 'derivation_id', 'min', 'max', 'median', 'footnote', 'min_year_acquired', 'gtin_upc']).sort_values('fdc_id')
```

```python
brnd_ntrn.head()
```

```python
brnd_ntrn_ntrn = pd.merge(brnd_ntrn, nutrient, left_on = 'nutrient_id', right_on = 'id').sort_values(['fdc_id', 'nutrient_id']).drop(columns = ['id', 'nutrient_nbr', 'rank'])
brnd_ntrn_ntrn.head()
```

```python
brnd_ntrn_ntrn = brnd_ntrn_ntrn.rename(columns = {'name': 'nutrient_name', 'amount': 'nutrient_amount', 'unit_name': 'nutrient_unit'})
brnd_ntrn_ntrn.head()
```

```python
brnd_ntrn_ntrn = brnd_ntrn_ntrn[['fdc_id', 'nutrient_id', 'nutrient_name', 'nutrient_amount', 'nutrient_unit', 'branded_food_category', 'brand_owner', 'serving_size', 'serving_size_unit', 'ingredients', 'household_serving_fulltext']]
brnd_ntrn_ntrn.head()
```

```python
brnd_energy = brnd_ntrn_ntrn[brnd_ntrn_ntrn['nutrient_name'] == 'Energy'].drop(columns = ['nutrient_id', 'nutrient_name'])
brnd_energy.head()
```

```python
brnd_energy.loc[(brnd_energy['nutrient_unit'] == 'kJ'), 'nutrient_amount'] /= 4.18585182085
```

```python
brnd_energy = brnd_energy.rename(columns = {'nutrient_amount': 'energy'})
brnd_energy.head()
```

Distribution of the energy of foods.

Some values are possibly incorrect.

```python
plt.hist(brnd_energy['energy'], bins = 1000)
plt.yscale('log')
```

 Since the energy of 100g oil is about 900kcal, we simply ignore rows with energy greater than 1000kcal.

 This does not excludes all the incorrect data. Detailed processing is still needed.

```python
brnd_energy_small = brnd_energy[brnd_energy['energy'] < 1000]
brnd_energy_small.sort_values('energy', ascending = False).head()
```

Which types of food have highest energy?

```python
cate_energy = brnd_energy.groupby('branded_food_category').agg({'energy': 'mean', 'fdc_id': 'count'})
cate_energy = cate_energy[cate_energy['fdc_id'] > 100]
cate_energy.sort_values('energy', ascending = False).head(10)
```

Which companies produce high-energy foods? (Or simply reporting incorrect data?)

Brand owners with more than 100 products will be counted.

```python
cate_bo_energy = brnd_energy.groupby('brand_owner').agg({'energy': 'mean', 'fdc_id': 'count'})
cate_bo_energy[cate_bo_energy['fdc_id'] > 100].sort_values('energy', ascending = False).head(10)
```

We can see that companies producing chocolates, nuts and fried food are in this list, which is reasonable.

Abbott data is possibly incorrect (possbily due to wrong unit), further check is needed.

### 3. Diets

```python
diets = pd.read_csv(os.path.join(diets_path, "All_Diets.csv"))
diets.head()
```

```python
diets_grouped = diets.groupby('Diet_type').agg({'Cuisine_type': 'count', 'Protein(g)': 'mean', 'Carbs(g)': 'mean', 'Fat(g)': 'mean'})
diets_grouped
```

```python
cuisine_grouped = diets.groupby('Cuisine_type').agg({'Cuisine_type': 'count', 'Protein(g)': 'mean', 'Carbs(g)': 'mean', 'Fat(g)': 'mean'})
cuisine_grouped
```

```python
diets_to_cuisine = diets.groupby(['Diet_type', 'Cuisine_type']).agg({'Cuisine_type': 'count', 'Protein(g)': 'mean', 'Carbs(g)': 'mean', 'Fat(g)': 'mean'})
diets_to_cuisine[diets_to_cuisine['Cuisine_type'] > 20].head(50)
```
